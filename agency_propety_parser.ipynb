```python
def price_parser(x):
    price = re.findall('\$([\d,]+)',x)
    if price:
        return int(min(price).replace(',',''))  # remove the comma from the price
    else:
        return 0

df['price'] = df['displayPrice'].apply(price_parser)

df['postcode'] = df['address'].apply(lambda x: x['postcode'])
df['streetAddress'] = df['address'].apply(lambda x: x['streetAddress'])
df['suburb'] = df['address'].apply(lambda x: x['suburb'])

df['bedrooms'] = df['features'].apply(lambda x:x['bedrooms'])
df['bathrooms'] = df['features'].apply(lambda x:x['bathrooms'])
df['parkingSpaces'] = df['features'].apply(lambda x:x['parkingSpaces'])
```
